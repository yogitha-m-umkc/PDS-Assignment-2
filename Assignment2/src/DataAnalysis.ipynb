```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./clean_data.csv")
df
```

```python
from matplotlib import pyplot as plt
df['Year'].plot(kind='hist', bins=20, title='Year')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Pie chart of transmission types
transmission_counts = df['Transmission'].value_counts()
plt.figure(figsize=(8, 6))
transmission_counts.plot(kind='pie', autopct="%1.1f%%", colors=['gold', 'yellowgreen'])
plt.title('Pie Chart of Transmission Types')
plt.xlabel('')
plt.ylabel('')
plt.axis('equal')
plt.legend(transmission_counts.index, loc='upper left')
plt.show()
```

```python
#bar chart of fuel types
fuel_type_counts = df['Fuel_Type'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(fuel_type_counts.index, fuel_type_counts.values, color=['gold', 'black', 'violet', 'red'])
plt.xlabel('Fuel Type')
plt.ylabel('Number of Vehicles')
plt.title('Count of Vehicles by Fuel Type')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Question C
df = pd.get_dummies(df, columns=['Fuel_Type', 'Transmission'])
```

```python
df
```

```python
# Question D
from datetime import datetime
# Calculate current year
current_year = datetime.now().year

# Calculate the age of the car
df['Car_Age'] = current_year - df['Year']
```

```python
df.head()
```

```python
# Defining function to categorize engine capacity
def categorize_engine_capacity(engine):
    if engine <= 1390:
        return 'Small'
    elif 1390 < engine <= 1950:
        return 'Medium'
    else:
        return 'Big'

df['Engine_Capacity'] = df['Engine'].apply(categorize_engine_capacity)
```

```python
df.head()
```

```python
# Counting the occurrences of each category
engine_capacity_counts = df['Engine_Capacity'].value_counts()
engine_capacity = [1.0, 1.2, 1.5, 1.8, 2.0, 2.2, 2.5]
engine_capacity_counts = [3, 5, 7, 2, 4, 1, 1]
plt.figure(figsize=(8, 6))
plt.plot(engine_capacity, engine_capacity_counts, marker='o', linestyle='-', color='skyblue')
plt.title('Line chart for Distribution of Engine Capacity')
plt.xlabel('Engine Capacity')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#Box Plot chart
plt.figure(figsize=(8, 6))
plt.boxplot(engine_capacity_counts, vert=False)
plt.title('BoxPlot for Distribution of Engine Capacity')
plt.xlabel('Engine Capacity')
plt.tight_layout()
plt.show()
```

```python
# Question E
# Select: Choose specific columns
Picked_columns = df[['Name', 'Location', 'Year', 'Price']]

print("Picked Columns:")
print(Picked_columns)
```

```python
filtered_data = df[df['Fuel_Type_Diesel'] == 1]

num_rows_filtered_data = len(filtered_data)


print("\nFiltered Data (Diesel cars only):")
print("Number of rows in the filtered data:", num_rows_filtered_data)
print(filtered_data)
```

```python
filtered_data = df[df['Engine_Capacity'] == 'Medium']

num_rows_filtered_data = len(filtered_data)

print("\nFiltered Data :")
print("Number of rows in the filtered data:", num_rows_filtered_data)
print(filtered_data)
```

```python
# Rename: Rename columns
renamed_df = df.rename(columns={'Unnamed: 0': 'ID', 'Kilometers_Driven': 'Km_Driven'})

print("\nRenamed Columns:")
print(renamed_df)
```

```python
df['Price_in_lakhs'] = df['Price'] * 100000

print("\nMutated DataFrame with Price in Lakh:")
print(df)
```

```python
# Scatter plot of Engine vs. Price
plt.figure(figsize=(10, 6))
plt.scatter(df['Engine'], df['Price'], color='maroon', alpha=0.5)
plt.xlabel('Engine (CC)')
plt.ylabel('Price (in Lakhs)')
plt.title('Scatter Plot of Engine vs. Price')
plt.grid(True)
plt.show()
```

```python
# Arrange: Sort rows by one or more columns
sorted_yr = df.sort_values(by='Year', ascending=False)

print("\nSorted DataFrame by Year:")
print(sorted_yr)
```

```python
# Arrange: Sort rows by one or more columns
sorted_Mileage = df.sort_values(by='Mileage', ascending=False)

print("\nSorted DataFrame by Mileage of the car:")
print(sorted_Mileage)
```

```python
# Group by operations
grouped_summary = df.groupby('Location').agg({'Price': 'mean', 'Year': 'max'}).reset_index()
print("\nGrouped Summary by Location:")
print(grouped_summary)
```

```python
grouped_data = df.groupby(['Location', 'Year']).size().reset_index(name='Count')

print(grouped_data)
```

```python
import matplotlib.pyplot as plt
grouped_cars_by_location = df.groupby('Location').size().reset_index(name='Number_of_Cars')

print(grouped_cars_by_location)
```

```python
#bar chart for number of cars by location
plt.figure(figsize=(8, 6))
plt.bar(grouped_cars_by_location['Location'], grouped_cars_by_location['Number_of_Cars'], color='purple')
plt.xlabel('Location')
plt.ylabel('Number of Cars')
plt.title('Bar chart for Number of Cars by Location')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Group by car age
grouped_data = df.groupby('Car_Age').size().reset_index(name='Cars_Count')

print("Grouped data by car age:")
print(grouped_data)
```

```python
# Plotting using Matplotlib
plt.figure(figsize=(10, 6))
plt.bar(grouped_data['Car_Age'], grouped_data['Cars_Count'], color='brown')
plt.xlabel('Car Age')
plt.ylabel('Number of Cars')
plt.title('Number of Cars by Car Age')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
